# Compute and save 1d Movement Signal for all ROIs for both participants

```python
directory = os.getcwd() + "/"
hdf5er = np.array([x for x in os.listdir(directory)])[np.array([".hdf5" in x for x in os.listdir(directory)])]
jsons = np.array([x for x in os.listdir(directory)])[np.array([".json" in x for x in os.listdir(directory)])]

outfolder = directory
```

```python
#For all hdf5, for all head hands and body jsons compute weighted histograms if input was present
mot = {}
one_d = {}
for file in hdf5er:
    root = re.search("(.*)\.",file).groups()[0]
    identifier = [root+"_left_head",root+"_left_hands",root+"_left_body",
                 root+"_right_head",root+"_right_hands",root+"_right_body"]
    jsons = [root+"_left_head.json",root+"_left_hands.json",root+"_left_body.json",
            root+"_right_head.json",root+"_right_hands.json",root+"_right_body.json"]
    for i,j in zip(identifier,jsons):
        try:
            with open(directory+j, encoding='utf-8') as fh:
                roi = json.load(fh)
                h264 = tables.open_file(directory+file, mode='r')
                h264 = h264.root.motion_tensor
                mothistmap = helper.weighted_histograms(h264,roi=roi)
                mot[i] = mothistmap#append mothistmap
                
                hist = ndimage.gaussian_filter(mothistmap,10)
                amount = np.sum(hist,axis=0)#1d signal
                one_d[i] = amount
        except Exception as e:
            pass
```

```python
with open(outfolder+"1d_movement.motion","wb") as f:
    pickle.dump(one_d,f)
```

```python
with open(outfolder+"2d_movement.motion","wb") as f:
    pickle.dump(mot,f)
```

```python
with open(outfolder+"1d_movement.motion","rb") as f:
    one_d = pickle.load(f)
```

# Compute contrast between all electrodes in alpha, beta and gamma range

```python
import os
import re
import numpy as np
from scipy import ndimage

import sys
import pickle
import json
import tables

import importlib.util #import from superordinate folder
spec = importlib.util.spec_from_file_location("helper", 
                                              str(re.search("(.*/)",os.getcwd()).groups(0)[0])+"helper.py")
helper = importlib.util.module_from_spec(spec)
spec.loader.exec_module(helper)
```

```python
folder = "/mnt/c/Users/eler/Documents/Cognitive_Science/Bachelorarbeit/Scripts/processing_pipeline/dyad_13/"
path = folder+"DualEEG_RPS_C_13.eeg"
vmrk = folder+"DualEEG_RPS_C_13.vmrk"
```

```python
def frequency_bands_between_electrodes(path, participant = 0, vmrk = None):
    time = 0
    if vmrk:
        idx = [x == "S128" for x in helper.parse_vmrk(vmrk)['description']].index(True)
        time = int(helper.parse_vmrk(vmrk)['time'][idx])//20
    
    alpha_contrasts = np.ndarray((32,32),dtype=np.object)
    beta_contrasts = np.ndarray((32,32),dtype=np.object)
    gamma_low_contrasts = np.ndarray((32,32),dtype=np.object)
    gamma_high_contrasts = np.ndarray((32,32),dtype=np.object)

    freq_alpha = [8,14]
    freq_beta = [14,30]
    freq_gamma_low = [30,50]
    freq_gamma_high = [50,100]

    participant = participant#0 or 1

    i = 1
    for electrode_1 in range(32):
        for electrode_2 in range(32):
            if(electrode_1 < electrode_2):
                continue
            contrast = helper.load_eeg(path,(participant*32)+electrode_1)-helper.load_eeg(path,(participant*32)+electrode_2)
            spectrum = np.abs(helper.spectrogram(contrast, f_min=0,f_max=100))
            print((1/((32*32)/2))*i)
            alpha_contrasts[electrode_1][electrode_2] = np.sum(spectrum[freq_alpha[0]:freq_alpha[1],time:],axis=0)
            beta_contrasts[electrode_1][electrode_2] = np.sum(spectrum[freq_beta[0]:freq_beta[1],time:],axis=0)
            gamma_low_contrasts[electrode_1][electrode_2] = np.sum(spectrum[freq_gamma_low[0]:freq_gamma_low[1],time:],axis=0)
            gamma_high_contrasts[electrode_1][electrode_2] = np.sum(spectrum[freq_gamma_high[0]:freq_gamma_high[1],time:],axis=0)
            i+=1
    return np.array([alpha_contrasts, beta_contrasts, gamma_low_contrasts, gamma_high_contrasts])
```

```python
data = frequency_bands_between_electrodes(path=path, participant = 0, vmrk =vmrk)
```
