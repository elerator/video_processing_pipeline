```python
from collections import defaultdict

corrcoef = defaultdict(lambda: defaultdict(dict))
n_larger_point_two = defaultdict(lambda: defaultdict(dict))
n_sig = defaultdict(lambda: defaultdict(dict))
avg_sig_bonf = defaultdict(lambda: defaultdict(dict))


both = defaultdict(lambda: defaultdict(dict))
```

```python
def print_table(dataframe, heading = " ", heading_indent = 4, subtext =""):
    heading = heading.replace("\\t","&nbsp;&nbsp;&nbsp;&nbsp;")
    heading = heading.replace(" ","&nbsp;")
    subtext = subtext.replace("\\t","&nbsp;&nbsp;&nbsp;&nbsp;")
    subtext = subtext.replace("\\n","<br>")


    if str(type(dataframe))=="<class 'collections.defaultdict'>":
        for x in dataframe.values():
            if str(type(x)) == "<class 'collections.defaultdict'>":
                dataframe = df_from_nested_dict(dataframe)
                break
    if str(type(dataframe))=="<class 'collections.defaultdict'>":
        dataframe = pd.DataFrame.from_dict(dataframe)

    dataframe = pd.DataFrame(dataframe)
    
    html_string = dataframe.to_html()
    
    
    html_string = html_string.replace("\\t","&nbsp;&nbsp;&nbsp;&nbsp;")
    html_string = html_string.replace("\\n","<br>")
    html_string = html_string.replace('<table border="1" class="dataframe">', '<table border="{}" cellpadding="{}" cellspacing="0" class="dataframe">'.format(10,10))
    

    
    display(HTML(  '<html>'
                 + '<h4 style="color: DarkBlue;text-align: left; padding-left:{0:3}px;">'.format(heading_indent)
                 + heading
                 + '</h4>'
                 + html_string
                 + '<div style="font-size: 12px">'
                 + subtext
                 + "</div>"
                 + "</html>"))

def df_from_nested_dict(d):
    return pd.DataFrame.from_dict({(i,j): d[i][j] for i in d.keys() for j in d[i].keys()},orient='index')
```

```python
subject = "RPS_C_13_C3_left_"

trial_periods = [[2200,8200],[10000,16000]]#Select a large portion of within trial regions
mixed = [0,12000]#Have the same n for a mixed period


one_d = one_d

names = {0:"Alpha Power", 1:"Beta Power", 2:"Low Gamma Power", 3:"High Gamma Power", "hands": "Hand Region", "body":"Body Region", "head": "Head Region",
        "complete": "Complete condition", "trial":"Trial period only"}

for part in ["complete", "trial"]:
    for frequency_band in [0,1,2,3]:
        for roi in ["hands","body","head"]:
            crosscor = None
            sig = None
            if part == "complete":
                sample = select_period(bands,frequency_band,mixed[0],mixed[1])
                motion_mixed = np.array(list(one_d[subject+roi][mixed[0]:mixed[1]]))
                corr_results = helper.correlations(sample,motion_mixed,smoothing = True)
            elif part == "trial":
                band_trial_only = list(select_period(bands,frequency_band,trial_periods[0][0],trial_periods[0][1]))
                band_trial_only.extend(list(select_period(bands,frequency_band,trial_periods[1][0],trial_periods[1][1])))
                band_trial_only = np.array(band_trial_only)
                
                motion_trial = list(np.array(list(one_d[subject+roi][trial_periods[0][0]:trial_periods[0][1]])))
                motion_trial.extend(list(one_d[subject+roi][trial_periods[1][0]:trial_periods[1][1]]))
                motion_trial = np.array(motion_trial)
                corr_results = helper.correlations(band_trial_only,motion_trial,smoothing = True)
            
            #Prepare crosscorrelation matrix
            crosscor = corr_results[0][:][:]
            crosscor = crosscor.flatten()
            
            #Prepare significance table by filling upper triangle and setting diagonal to nan
            sig = corr_results[1][:][:]
            sig = sig.T+sig
            np.fill_diagonal(sig, np.nan)
            #Comute average corrected significance
            mean_p = np.nanmean(sig)
            avg_sig_bonf[names[part]][names[frequency_band]][names[roi]]= mean_p

            
            
            #Calculate mean, standard deviation the number of correlations larger .3 and the number of corrected significant values
            mean = np.nansum(np.abs(crosscor))/(np.count_nonzero(~np.isnan(crosscor))+32)
            std = np.nanstd(crosscor)
            nl = len(np.argwhere(np.abs(crosscor)>.3))//2
            ns = len(np.argwhere(corrected_sig <.05))#number of significant values corrected
            
            #print(bonferroni_sig)
            #[names[part]][names[frequency_band]][names[roi]] = np.nanmean(bonferroni_sig)
            corrcoef[names[part]][names[frequency_band]][names[roi]]= "{0:1.2f}".format(mean)
            n_larger_point_two[names[part]][names[frequency_band]][names[roi]]= "{0:1}".format(nl)
            n_sig[names[part]][names[frequency_band]][names[roi]]= ns
            
            if mean_p < .05:
                both[names[part]][names[frequency_band]][names[roi]]= "\t" +str(nl) + " (r = "+"{0:1.2f}".format(mean) +"*)"
            elif mean_p < .01:
                both[names[part]][names[frequency_band]][names[roi]]= "\t" +str(nl) + " (r = "+"{0:1.2f}".format(mean) +"**)"
            else:
                both[names[part]][names[frequency_band]][names[roi]]= "\t" +str(nl) + " (r = "+"{0:1.2f}".format(mean) +")"
```
